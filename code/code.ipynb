Functions to plot the custom decision tree graph to visualise the rules output

```python
import igraph
from igraph import Graph, EdgeSeq
import plotly.graph_objects as go

def treeplot(vertices, edges, Labels):
  nr_vertices = vertices
  v_label = list(map(str, range(nr_vertices)))
  # G = Graph.Tree(nr_vertices, 2) # 2 stands for children number
  g = Graph()
  g.add_vertices(nr_vertices)
  g.add_edges(edges)
  lay = g.layout_reingold_tilford(root=[0])

  position = {k: lay[k] for k in range(nr_vertices)}
  Y = [lay[k][1] for k in range(nr_vertices)]
  M = max(Y)

  es = EdgeSeq(g) # sequence of edges
  E = [e.tuple for e in g.es] # list of edges

  L = len(position)
  Xn = [position[k][0] for k in range(L)]
  Yn = [2*M-position[k][1] for k in range(L)]
  Xe = []
  Ye = []
  for edge in E:
    Xe+=[position[edge[0]][0],position[edge[1]][0], None]
    Ye+=[2*M-position[edge[0]][1],2*M-position[edge[1]][1], None]

  labels = Labels
  fig = go.Figure()
  fig.add_trace(go.Scatter(x=Xe,
                    y=Ye,
                    mode='lines',
                    line=dict(color='rgb(210,210,210)', width=1),
                    hoverinfo='none'
                    ))
  fig.add_trace(go.Scatter(x=Xn,
                    y=Yn,
                    mode='markers',
                    name='bla',
                    marker=dict(symbol='circle-dot',
                                  size=60,
                                  color='#6175c1',    #'#DB4551',
                                  line=dict(color='rgb(50,50,50)', width=1)
                                  ),
                    text=labels,
                    hoverinfo='text',
                    opacity=0.8
                    ))
  axis = dict(showline=False, # hide axis line, grid, ticklabels and  title
              zeroline=False,
              showgrid=False,
              showticklabels=False,
              )

  fig.update_layout(title= 'Decision Tree',
                annotations=make_annotations(position, v_label, labels, M, position),
                font_size=12,
                showlegend=False,
                xaxis=axis,
                yaxis=axis,
                margin=dict(l=40, r=40, b=85, t=100),
                hovermode='closest',
                plot_bgcolor='rgb(248,248,248)'
                )
  fig.show()
```

```python
def make_annotations(pos, text, labels, M, position, font_size=8, font_color='rgb(0,0,0)'):
    L=len(pos)
    if len(text)!=L:
        raise ValueError('The lists pos and text must have the same len')
    annotations = []
    for k in range(L):
        annotations.append(
            dict(
                text=labels[k], # or replace labels with a different list for the text within the circle
                x=pos[k][0], y=2*M-position[k][1],
                xref='x1', yref='y1',
                font=dict(color=font_color, size=font_size),
                showarrow=False)
        )
    return annotations
```

```python
def print_decision_tree(node, columns, features): 
  vertices = 0
  labels = list()
  edges = list()
  node['node_index'] = 0
  que = [node]
  while len(que) > 0:
    nd = que.pop(0)
    if isinstance(nd, dict):
      labels.append(get_node_text(nd, columns, features))
      vertices += 1
      if isinstance(nd['left'], dict):
        nd['left']['node_index'] = vertices
        edges.append([nd['node_index'],  nd['left']['node_index']])
      else:
        edges.append([nd['node_index'], vertices])
      que.append(nd['left'])
      vertices += 1
      if isinstance(nd['right'], dict):
        nd['right']['node_index'] = vertices
        edges.append([nd['node_index'],  nd['right']['node_index']])
      else:
        edges.append([nd['node_index'], vertices])
      que.append(nd['right'])
    else:
      labels.append(str(nd[0]))
  vertices += 1
  treeplot(vertices, edges, labels)

```

```python
def get_node_text(node, columns, features):
  text = ""
  if features ==2:
    text = columns[node['index'][0]] + "(w1) " + columns[node['index'][1]] + "(w2) <br>" + "%.4f"%(node['value'][0]) + "w1 + " + "%.4f"%(node['value'][1]) + "w2 < " + "%.4f"%(node['value'][2])
  else:
    text = columns[node['index']] + "<br>" + "Value: " + "%.4f"%(node['value']) #refinement float points
  return text
```

Function to Print AUC ROC CURVE

```python
from sklearn.metrics import roc_curve
def plot_AOC_ROC_Curve(classes, y_pred_proba, y_test, title):
  labels = ['Class 1', 'Class 2', 'Class 3']
  colors = ['orange', 'green', 'red']
  random_probs = [0 for i in range(len(y_test))]
  p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
  plt.plot(p_fpr, p_tpr, linestyle='--',color='cyan')
  if classes == 1:
    fpr, tpr, thresh = roc_curve(y_test, y_pred_proba[:,1], pos_label=1)
    plt.plot(fpr, tpr, linestyle='--',color=colors[0], label=labels[0])
    plt.title(title)
  else:
    fpr = {}
    tpr = {}
    thresh = {}
    for i in range(1, classes+1):
        fpr[i-1], tpr[i-1], thresh[i-1] = roc_curve(y_test, y_pred_proba[:,i-1], pos_label=i) # Calculating fpr and tpr for plotting the ROC AUC Curve
    for i in range(classes):
      print(fpr[i].shape)
      plt.plot(fpr[i], tpr[i], linestyle='--',color=colors[i], label=labels[i])
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive rate')
  plt.title(title)
  plt.legend(bbox_to_anchor=(1,1), loc=2)
  plt.show()
```

Custom Decision Tree Classifier with Logistic regression to split the node 

```python
from sklearn.linear_model import LogisticRegression
from itertools import combinations
import random
import math
from sklearn.metrics import accuracy_score 
class LogisticDecisionTrees():
  def __init__(self, max_depth, min_samples, max_features, n_features):
    self.max_depth = max_depth
    self.min_samples = min_samples
    self.max_features = max_features
    self.n_features = n_features
  
  def calculate_entropy(self, data):
    label_column = data[1]
    _, counts = np.unique(label_column, return_counts=True)
    probabilities = counts / counts.sum()
    entropy = sum(probabilities * -np.log2(probabilities))
    return entropy

  def calculate_overall_entropy(self, split_1, split_2):
    n = len(split_1[0]) + len(split_2[0])
    p1 = len(split_1[0]) / n
    p2 = len(split_2[0]) / n
    overall_entropy =  (p1 * self.calculate_entropy(split_1)  + p2 * self.calculate_entropy(split_2))
    return overall_entropy

  def determine_best_split(self, data, features, values):
    overall_entropy = 9999
    if self.n_features == 2:
      for column in range(len(features)):
        split_1, split_2 = self.test_split_pair(features[column], values[column], data[0], data[1])
        current_overall_entropy = self.calculate_overall_entropy(split_1, split_2)
        if current_overall_entropy <= overall_entropy:
          overall_entropy = current_overall_entropy
          feature = features[column]
          value = values[column]
    else:
      for column in range(len(features)):
        split_1, split_2 = self.test_split(features[column], values[column], data[0], data[1])
        current_overall_entropy = self.calculate_overall_entropy(split_1, split_2)
        if current_overall_entropy <= overall_entropy:
          overall_entropy = current_overall_entropy
          feature = features[column]
          value = values[column]
    return feature, value

  def test_split(self, index, value, X, Y):
    left, right = list(), list()
    lp, rp = list(), list()
    for i in range(len(X)):
      if X[i][index] < value:
        left.append(X[i])
        lp.append(Y[i])
      else:
        right.append(X[i])
        rp.append(Y[i])
      l = [left, lp]
      r = [right, rp]
    return [l, r]

  def test_split_pair(self, index, value, X, Y):
    left, right = list(), list()
    lp, rp = list(), list()
    for i in range(len(X)):
      val = X[i][index[0]]*value[0] + X[i][index[1]]*value[1]
      if val < value[2]:
        left.append(X[i])
        lp.append(Y[i])
      else:
        right.append(X[i])
        rp.append(Y[i])
      l = [left, lp]
      r = [right, rp]
    return [l, r]

  def get_split_pair(self, dataset):
    feature_list = random.sample(range(0, len(dataset[0][0])), self.max_features)
    feature_pairs = list(combinations(feature_list, 2))
    values = list()
    features_pairs = list()
    y = len(set(dataset[1]))
    flag = False
    if y > 2:
      flag = True
    for feature in feature_pairs:
      f1 = feature[0]
      f2 = feature[1]
      X = list()
      for row in dataset[0]:
        X.append([row[f1], row[f2]])
      Y = dataset[1]
      logclf = LogisticRegression(max_iter = 2000, random_state = 1)
      logclf.fit(X, Y)
      if flag:
        for i in range(y):
          features_pairs.append(feature)
          values.append([logclf.coef_[i][0], logclf.coef_[i][1], -logclf.intercept_[i]])
      else:
        values.append([logclf.coef_[0][0], logclf.coef_[0][1], -logclf.intercept_[0]])
        features_pairs.append(feature)
    features, value = self.determine_best_split(dataset, feature_pairs, values)
    groups = self.test_split_pair(features, value, dataset[0], dataset[1])
    return {'index':features, 'value':value, 'groups':groups}
  
  def get_split_single(self, dataset):
    feature_list = random.sample(range(0, len(dataset[0][0])), self.max_features)
    predictions = list()
    values = list()
    features_list = list()
    y = len(set(dataset[1]))
    flag = False
    if y > 2:
      flag = True
    for feature in feature_list:
      X = list()
      for row in dataset[0]:
        X.append([row[feature]])
      Y = dataset[1]
      logclf = LogisticRegression(max_iter = 2000, random_state = 1)
      logclf.fit(X, Y)
      if flag:
        for i in range(y):
          features_list.append(feature)
          if logclf.coef_[i][0] == 0:
            values.append((-1 * (np.sign(logclf.intercept_[i])) * math.inf))
          else:
            values.append((-logclf.intercept_[i]/logclf.coef_[i][0]))
      else:
        if logclf.coef_[0][0] == 0:
          values.append((-1 * (np.sign(logclf.intercept_[0])) * math.inf))
        else:
          values.append((-logclf.intercept_[0]/logclf.coef_[0][0]))
        features_list.append(feature)
    index, value = self.determine_best_split(dataset, features_list, values)
    groups = self.test_split(index, value, dataset[0], dataset[1])
    return {'index':index, 'value':value, 'groups':groups}


  def get_split(self, dataset):
    if self.n_features == 1:
      return self.get_split_single(dataset)
    else:
      return self.get_split_pair(dataset)

  def to_terminal(self, group):
    outcomes = group[1]
    n = len(outcomes)
    probs = []
    for i in sorted(self.unique_vals):
      probs.append(outcomes.count(i))
    probs = [i/n for i in probs]
    return [max(set(outcomes), key=outcomes.count), probs]

  def split(self, node, depth):
    left, right = node['groups'][0], node['groups'][1]
    del(node['groups'])
    # check for a no split
    if not left[0] or not right[0]:
      left[0].extend(right[0])
      left[1].extend(right[1])
      node['left'] = node['right'] = self.to_terminal(left)
      return
    # check for max depth
    if depth >= self.max_depth:
      node['left'], node['right'] = self.to_terminal(left), self.to_terminal(right)
      return
    # process left child
    if len(left[0]) <= self.min_samples:
      node['left'] = self.to_terminal(left)
    elif len(set(left[1])) == 1:
      node['left'] = self.to_terminal(left)
    else:
      node['left'] = self.get_split(left)
      self.split(node['left'], depth+1)
    # process right child
    if len(right[0]) <= self.min_samples:
      node['right'] = self.to_terminal(right)
    elif len(set(right[1])) == 1:
      node['right'] = self.to_terminal(right)
    else:
      node['right'] = self.get_split(right)
      self.split(node['right'], depth+1)

  def fit(self, X, Y):
    self.unique_vals = []
    for i in set(Y):
      self.unique_vals.append(i)
    self.unique_vals.sort()
    self.max_depth = min(X.shape[1], self.max_depth)
    self.root = self.get_split([X, Y])
    self.split(self.root, 1)

  def predict_sample_single(self, node, row):
    if row[node['index']] < node['value']:
      if isinstance(node['left'], dict):
        return self.predict_sample_single(node['left'], row)
      else:
        return node['left'][0]
    else:
      if isinstance(node['right'], dict):
        return self.predict_sample_single(node['right'], row)
      else:
        return node['right'][0]

  def predict_sample_pair(self, node, row):
    val = row[node['index'][0]]*node['value'][0] + row[node['index'][1]]*node['value'][1]
    if val < node['value'][2]:
      if isinstance(node['left'], dict):
        return self.predict_sample_pair(node['left'], row)
      else:
        return node['left'][0]
    else:
      if isinstance(node['right'], dict):
        return self.predict_sample_pair(node['right'], row)
      else:
        return node['right'][0]

  def predict(self, X):
    predictions = []
    if self.n_features == 1:
      for row in X:
        predictions.append(self.predict_sample_single(self.root, row))
    else:
      for row in X:
        predictions.append(self.predict_sample_pair(self.root, row))
    return predictions

  def predict_proba_sample_single(self, node, row):
    if row[node['index']] < node['value']:
      if isinstance(node['left'], dict):
        return self.predict_proba_sample_single(node['left'], row)
      else:
        return node['left'][1]
    else:
      if isinstance(node['right'], dict):
        return self.predict_proba_sample_single(node['right'], row)
      else:
        return node['right'][1]

  def predict_proba_sample_pair(self, node, row):
    val = row[node['index'][0]]*node['value'][0] + row[node['index'][1]]*node['value'][1]
    if val < node['value'][2]:
      if isinstance(node['left'], dict):
        return self.predict_proba_sample_pair(node['left'], row)
      else:
        return node['left'][1]
    else:
      if isinstance(node['right'], dict):
        return self.predict_proba_sample_pair(node['right'], row)
      else:
        return node['right'][1]

  def predict_proba(self, X):
    predictions = []
    if self.n_features == 1:
      for row in X:
        predictions.append(self.predict_proba_sample_single(self.root, row))
    else:
      for row in X:
        predictions.append(self.predict_proba_sample_pair(self.root, row))
    return predictions
```

5 Fold Cross Validation

```python
from random import seed
from random import randrange
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def cross_validation_split(dataset, folds):
	fold_size = int(len(dataset) / folds)
	dataset_split = np.empty(shape = (0, fold_size, dataset.shape[1]))
	dataset_copy = np.array(dataset)
	for i in range(folds):
		fold = np.empty(shape = (0, dataset.shape[1]))
		for j in range(fold_size):
			index = randrange(len(dataset_copy))
			fold = np.insert(fold, j, dataset_copy[index, :], axis = 0)
			dataset_copy = np.delete(dataset_copy, index, axis=0)
		dataset_split = np.insert(dataset_split, i, fold, axis = 0)
	return dataset_split

seed(1)
acc = []
fscore = []
prec = []
recall = []
folds = cross_validation_split(dfcancer_data, 5)
train_sets = []
test_sets = []
models_list = []
datasets = [cancer_data, fetal_data, bank_data]
labels_data = ["Cervical Cancer Risk", "Fetal Health", "Banking"]
labels_att = ["1 Attribute for Split", "2 Attibutes for Split"]

for dataset in datasets:
  acc1 = []
  acc2 = []
  p1 = []
  p2 = []
  f1s = []
  f1s2 = []
  rec1 = []
  rec2 = []
  folds = cross_validation_split(dataset, 5)
  for i in range(5): #Loop to perform K experiments
    test_sets.append(folds[i])
    train_set = None
    initindex = 0
    for j in range(5):
      if(j != i):
        train_set = folds[j]
        initindex = j
        break
    for j in range(initindex+1, 5):
      if j != i:
        train_set = np.append(train_set, folds[j], axis = 0)
    train_sets.append(train_set)
    logit = LogisticDecisionTrees(max_depth = 5, min_samples = 1, max_features = train_sets[i].shape[1] - 1, n_features = 1)
    logit2 = LogisticDecisionTrees(max_depth = 4, min_samples = 1, max_features = train_sets[i].shape[1] - 1, n_features = 2)
    logit.fit(train_sets[i][:, :-1], train_sets[i][:,-1])
    logit2.fit(train_sets[i][:, :-1], train_sets[i][:,-1])
    y_pred = logit.predict(test_sets[i][:,:-1])
    y_pred2 = logit2.predict(test_sets[i][:,:-1])
    acc1.append(accuracy_score(test_sets[i][:, -1], y_pred))
    p1.append(precision_score(test_sets[i][:, -1], y_pred, average = 'macro'))
    f1s.append(f1_score(test_sets[i][:, -1], y_pred, average = 'macro'))
    rec1.append(recall_score(test_sets[i][:, -1], y_pred, average = 'macro'))
    acc2.append(accuracy_score(test_sets[i][:, -1], y_pred2))
    p2.append(precision_score(test_sets[i][:, -1], y_pred2, average = 'macro'))
    f1s2.append(f1_score(test_sets[i][:, -1], y_pred2, average = 'macro'))
    rec2.append(recall_score(test_sets[i][:, -1], y_pred2, average = 'macro'))
  acc.append([np.array(acc1, dtype = np.float64).mean(), np.array(acc2, dtype = np.float64).mean()])
  fscore.append([np.array(f1s, dtype = np.float64).mean(), np.array(f1s2, dtype = np.float64).mean()])
  prec.append([np.array(p1, dtype = np.float64).mean(), np.array(p2, dtype = np.float64).mean()])
  recall.append([np.array(rec1, dtype = np.float64).mean(), np.array(rec2, dtype = np.float64).mean()])
for i in range(3):
  print(labels_data[i])
  for j in range(2):
    print(labels_att[j])
    print("Accuracy Score: %.16f"%(acc[i][j]))
    print("Recall Score: %.16f"%(recall[i][j]))
    print("F1 Score Score: %.16f"%(fscore[i][j]))
    print("Precision Score: %.16f"%(prec[i][j]))
```

1. Resampled Paired t-test

```python
import numpy as np
# Y = np.array(dffetal['fetal_health'], dtype = int)
# X = np.array(dffetal.drop('fetal_health', axis = 1), dtype = np.float64)
Y = np.array(dfbank['y'], dtype = int)
X = np.array(dfbank.drop('y', axis = 1), dtype = np.float64)
```

```python
import numpy as np
from sklearn.model_selection import KFold
from scipy.stats import t as t_dist
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

logit1 = LogisticDecisionTrees(max_depth = 5, min_samples = 1, max_features = 33, n_features = 1)
logit2 = LogisticDecisionTrees(max_depth = 4, min_samples = 1, max_features = 33, n_features = 2)

def paired_t_test(p):
    p_hat = np.mean(p)
    n = len(p)
    den = np.sqrt(sum([(diff - p_hat)**2 for diff in p]) / (n - 1))
    t = (p_hat * (n**(1/2))) / den
    
    p_value = t_dist.sf(t, n-1)*2
    
    return t, p_value

n_tests = 30
accu1 = []
accu2 = []
p_ = []
rng = np.random.RandomState(42)
for i in range(n_tests):
    randint = rng.randint(low=0, high=32767)
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=randint)
    logit1.fit(X_train, y_train)
    logit2.fit(X_train, y_train)
    acc1 = accuracy_score(y_test, logit1.predict(X_test))
    acc2 = accuracy_score(y_test, logit2.predict(X_test))
    accu1.append(acc1)
    accu2.append(acc2)
    p_.append(acc1 - acc2)
    
print("Paired t-test Resampled")
t, p = paired_t_test(p_)
print(f"t statistic: {t}, p-value: {p}\n")
print("One Attribute Split Accuracy: %.16f"%(np.array(accu1, dtype = np.float64).mean()))
print("Two Attributes Split Accuracy: %.16f"%(np.array(accu2, dtype = np.float64).mean()))
if p <= 0.05:
  print("The models might not work in the same way. The statistical difference might be significant. (H1 True)")
else:
  print("The models will work in the same way. (H0 True)")
```

2. Cross-Validated Paired t-test

```python
p_ = []
accu1 = []
accu2 = []
logit1 = LogisticDecisionTrees(max_depth = 5, min_samples = 1, max_features = 21, n_features = 1)
logit2 = LogisticDecisionTrees(max_depth = 4, min_samples = 1, max_features = 21, n_features = 2)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
for train_index, test_index in kf.split(X):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], Y[train_index], Y[test_index]
    logit1.fit(X_train, y_train)
    logit2.fit(X_train, y_train)
    acc1 = accuracy_score(y_test, logit1.predict(X_test))
    acc2 = accuracy_score(y_test, logit2.predict(X_test))
    accu1.append(acc1)
    accu2.append(acc2)
    p_.append(acc1 - acc2)

print("Cross Validated Paired t-test")
t, p = paired_t_test(p_)
print(f"t statistic: {t}, p-value: {p}\n")
print("One Attribute Split Accuracy: %.16f"%(np.array(accu1, dtype = np.float64).mean()))
print("Two Attributes Split Accuracy: %.16f"%(np.array(accu2, dtype = np.float64).mean()))
if p <= 0.05:
  print("The models might not work in the same way. The statistical difference might be significant. (H1 True)")
else:
  print("The models will work in the same way. (H0 True)")
```
